### つぎにいくまえに、、、

使いにくいデータを処理しておきましょう(解析の際に使えなさそうなところを修正)

```python
# 乗船場所はCSQが入っていて、このままでは数学的に〜
titanic['Embarked'] = titanic['Embarked'].replace(['C', 'S', 'Q'], [0,1,2])
```

```python
titanic
```

```python
# 性別、年齢も同じように修正する
titanic['Sex'] = titanic['Sex'].replace(['male', 'female'], [0,1])
titanic['Age'] = titanic['Age'].replace('NaN', 0)
```

```python
# 解析に不必要そうなもの、欠損値が多いもの等を削除する
titanic.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
```
